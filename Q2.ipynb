```python
import numpy as np
```

```python
class MaxNet():
    def __init__(self, n, e):
        self.n = n
        self.e = e
        self.init_W()
    
    def init_W(self):
        eye = np.eye(self.n, dtype='float')*self.e + np.eye(self.n, dtype='float')
        es = -1 * np.ones((self.n, self.n), dtype='float') * self.e
        self.W = eye + es

    def act_f(self, x):
        mask = 1*(x >= 0)
        x = x * mask
        return x

    def stop_cond(self, x):
        '''
        this stop condition is for cases that there is only one global maximum
        '''
        nzeros = np.sum(1*(x == 0))
        if nzeros == (self.n -1):
            return True
        return False

    def __call__(self, x, verbos=True):
        step = 0
        assert x.shape == (self.n, 1)
        if verbos:
                print(f'step({step}): {x.flatten()}')
        while not self.stop_cond(x):
            x = self.act_f(np.matmul(self.W, x))
            if verbos:
                step += 1
                print(f'step({step}): {x.flatten()}')
        return np.argmax(x)
```

```python
x = np.array([1.2, 1.1, 1, 0.9, 0.95, 1.15]).reshape(-1, 1)
epsilon = 0.15
net = MaxNet(x.shape[0], epsilon)
```
