```python
import numpy as np
```

```python
class HamNet():
    def __init__(self, n, bases):
        self.n_ = n
        self.m_ = len(bases)
        self.bases_ = bases
        self.init_W()
    
    def init_W(self):
        self.W = np.zeros((self.m_, self.n_+1))
        self.W[:, 0] = self.n_/2
        for i, base in enumerate(self.bases_):
            self.W[i, 1:] = base/2

    def act_f(self, x):
        return x

    def __call__(self, x):
        x = x.reshape(-1, 1)
        bias = np.ones((1,1))
        x = np.vstack((bias, x))
        x = self.act_f(np.matmul(self.W, x))
        return np.argmax(x)+1, x
```

```python
b1 = np.array([+1, -1, +1, -1, +1, -1])
b2 = np.array([-1, +1, -1, +1, -1, +1])
b3 = np.array([+1, +1, +1, +1, +1, +1])

v1  = np.array([+1, -1, +1, +1, -1, +1])
v2  = np.array([-1, +1, +1, -1, +1, -1])
v3  = np.array([+1, +1, +1, -1, -1, -1])
v4  = np.array([-1, -1, -1, +1, +1, +1])
v5  = np.array([+1, +1, +1, +1, +1, +1])
v6  = np.array([-1, -1, +1, -1, -1, -1])
v7  = np.array([-1, -1, -1, +1, -1, -1])
v8  = np.array([+1, +1, -1, -1, +1, +1])
v9  = np.array([+1, +1, -1, +1, +1, +1])
v10 = np.array([+1, +1, +1, -1, +1, +1])

bases = [b1, b2, b3]
vs = [v1, v2, v3, v4, v5, v6, v7, v8, v9 ,v10]
```

```python
net = HamNet(b1.shape[0], bases)
```

```python
for i, v in enumerate(vs):
    arg , dist = net(v)
    print(f'v{i+1} -> out = {dist.flatten()}, arg = {arg}')
```
